## Задание 8
>*8). Постройте матрицу корреляций для всех полей X. Дайте полученному датафрейму название X_corr.*

```python
# матрица корреляций строится путем вызова метода corr()
X_corr = X.corr()
X_corr
```

Попробуем визуализировать матрицу корреляций

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (15,10))

sns.set(font_scale=1.4)

corr_matrix = np.round(X_corr, 2)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Матрица корреляций')
plt.show()
```

## Задание 9
>*9). Создайте список high_corr из признаков, корреляция которых с полем target по абсолютному значению превышает 0.5 (причем, само поле target не должно входить в этот список).*

```python
# в матрице корреляций список признаков содержится в индексе.
high_corr = list(X_corr[(abs(X_corr.target) > 0.5) & (X_corr.index != 'target')].index)
print(f'Список признаков для которых коэфициент корреляции > 0.5: {high_corr}')
```

Попробуем визуализировать корреляцию признаков с целевой переменной

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (15,10))

sns.set(font_scale=1.4)

corr_matrix = np.round(X_corr[['target']], 2)

corr_matrix = corr_matrix[corr_matrix.index != 'target']
corr_matrix[np.abs(corr_matrix) < 0.5] = 0
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Корреляция признаков и целевой переменной')
plt.show()
```

## Задание 10
>*10). Удалите из датафрейма X поле с целевой переменной. Для всех признаков, названия которых содержатся в списке high_corr, вычислите квадрат их значений и добавьте в датафрейм X соответствующие поля с суффиксом '_2', добавленного к первоначальному названию признака. Итоговый датафрейм должен содержать все поля, которые, были в нем изначально, а также поля с признаками из списка high_corr, возведенными в квадрат. Выведите описание полей датафрейма X с помощью метода describe.*

```python
# удаляем bp датафрейма X целевую переменную (колонку) - target
X.drop(columns='target', inplace=True)
# возводим в квадрат значения полей, которые перечислены в списке high_corr
hs = X[high_corr]**2
# меняем названия колонок, добавляя суффикс "_2"
hs.columns = hs.columns + '_2'
# соединяем исходный датафрейм X и новый, с квадратами наиболее коррелированных признаков
# сохраняем в том же датафрейме
X = pd.concat([X, hs], axis=1)
# выводим информацию об объединенном датафрейме
X.info()
```
