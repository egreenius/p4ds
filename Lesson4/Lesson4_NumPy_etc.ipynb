## Задание 9
>*Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12. Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy), состоящая из 4 строк и 3 столбцов. Получить матрицу At путем транспонирования матрицы A. Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?*

```
a = np.arange(0, 12)
A = a.reshape(4, 3)
At = A.T
B = np.dot(A, At)
print(B)
print(f'Размер матрицы B = {B.size}, форма = {B.shape}, размерность = {B.ndim}')
# Обратная матрица - это матрица, при умножению на которую получаем единичную матрицу
print('Обратная матрица - это матрица, при умножению на которую получаем единичную матрицу')
# Алгоритм нахождения обратной матрицы
# 1. Определяем является ли наша матрица квадратной? Если не квадратная, то обратной матрицы не существует
print(f'Форма матрицы B = {B.shape}')
print('Матрица B - квадратная. Переходим к следующему шагу')
# 2. Считаем определитель матрицы. Если определитель = 0, то обратной матрицы не существует
print(f'Определитель матрицы B = {np.linalg.det(B)}')
print('Определитель матрицы B = 0. Стоп! - обратной матрицы, для матриы B - не существует!')
```

```
C = c.reshape(4, -1)
# умножение матрицы на скаляр (С*10) и сложение матриц поэлементно
D = B + C*10
print(f'Определитель матрицы D = {np.linalg.det(D)}')
print(f'Ранг матрицы D = {np.linalg.matrix_rank(D)}')
D_inv = np.linalg.inv(D)
print('Обратная матрица для матрицы D:')
# настроим немного печать матриц - если float это малое число, то будет печататься 0
# если параметр suppress=False (по умолчанию), то такие числа будут печататься в научном формате, с большой минусовой степенью
np.set_printoptions(suppress=True)
print(D_inv)
# проверяем - умножение матрицы на обратную должно дать единичную матрицу
print('Результат умножения матрицы D на обратную матрицу D_inv:')
print(D.dot(D_inv))
```

## Задание 13
>*Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице. Убедитесь, что в матрице D_inv остались только нули и единицы. С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных, получите матрицу E размером 4x4.  Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.*

```
# Условие в квадратных скобках является битовой маской, то есть запись D_inv[D_inv < 0] - 
# возвращает элементы матрицы D_inv, удовлетворяющие условию внутри квадратных скобок
D_inv[D_inv < 0] = 0
D_inv[D_inv > 0] = 1
print(D_inv)
```

```
# Матрица D_inv используется как маска в функции np.where: для позиций элементов, где удовлетворяется
# условие D_inv == 1, в итоговую матрицу берутся элементы из матрицы B, в позициях, где условие
# False в итоговую матрицу вставляются элементы из соответствующей позиции матрицы С
E = np.where(D_inv == 1, B, C)
print('Матрица D_inv:')
print(D_inv)
print('Матрица B:')
print(B)
print('Матрица C:')
print(C)
print('Матрица E:')
print(E)
```
