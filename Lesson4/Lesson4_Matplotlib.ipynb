Создаем массивв - t от 0 до 10

```python
t = np.linspace(0, 10, 51)
# print(t)
```

Рассчитываем массив как функцию от t (f = cos(t))

```python
# задаем словарь с описанием шрифта для заголовка
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "darkblue",
    "family": "colibri",
}
# задаем словарь с описанием шрифтка для названий осей
label_font = {
    "fontsize": 9,
    "family": "serif",
    "color": "darkblue",
}
# передаем массив данных и параметры отрисовки в функцию рисования
plt.plot(t, f, color="green")
# Определяем заголовок графика - текст, параметры шрифта, и  положение заголовка. 
#(loc="center" - по умолчанию, можно опустить)
plt.title('График f(t)', fontdict=title_font, loc='center')
# Определяем наименование по оси х
plt.xlabel("Значения t", fontdict=label_font)
# Определяем наименование по оси y
plt.ylabel("Значения f", fontdict=label_font)
# определяем границы по оси х и по оси y 
plt.axis([0.5, 9.5, -2.5, 2.5])
# Задаем параметр отображения сетки
plt.grid(color='lightgrey')
# Отображаем график
plt.show()
```

## Задание 3
>*С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.*

Задаем массив исходных данных

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x +0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
# Задаем фигуру, а также количество и расположение графиков на ней
fig, ax = plt.subplots(nrows=2, ncols=2)
# устанавливаем размеры фигуры в дюймах, и расстояния между графиками по вертикали и горизонтали
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
 
# Задаем параметры первого графика - Заголовок, границы по горизонтали, и рисуем график
ax[0][0].set_title('График y1')
ax[0][0].set_xlim(-5, 5)
ax[0][0].plot(x, y1)
# Задаем параметры второго графика 
ax[0][1].set_title('График y2')
ax[0][1].plot(x, y2)
# Задаем параметры третьего графика 
ax[1][0].set_title('График y3')
ax[1][0].plot(x, y3)
# Задаем параметры четвертого графика 
ax[1][1].set_title('График y4')
ax[1][1].plot(x, y4)
plt.show()
```

```python
import pandas as pd
```

```python
url = "creditcard.csv"
cc_info = pd.read_csv(url)
cc_Class_info = cc_info.Class.value_counts()
cc_Class_info
```

```python
# Setting FiveThirtyEight style
plt.style.use('fivethirtyeight')
```

```python
# строим обычную столбчатую диаграмму
cc_Class_info.plot(kind="bar")
plt.show()
```

```python
# задаем словарь с описанием шрифта для заголовка
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "darkblue",
    "family": "colibri",
}
# с помощью параметра logy = True задаем логарифмический масштаб по оси y 
cc_Class_info.plot(kind="bar", logy=True)
plt.title('Credit Card Class info', fontdict=title_font, loc='center')
plt.show()
```

```python
# создаем два набора данных для параметра V1: для обычных транзакций и для мошеннических
cc_info_V1_Class0 = cc_info[cc_info.Class == 0].V1
cc_info_V1_Class1 = cc_info[cc_info.Class == 1].V1

# подготавливаем полотно (axes) для рисования на нем гистограмм
_, ax = plt.subplots()
# задаем параметры гистограмм и строим графики
ax.hist(cc_info_V1_Class0, bins = 20, color = 'gray', alpha = 0.5, label = 'Class0', density=True,)
ax.hist(cc_info_V1_Class1, bins = 20, color = 'red', alpha = 0.5, label = 'Class1', density=True,)
#ax.set_ylabel(y_label)
# не понимаю - зачем задавать подпись оси Х - Class?!
ax.set_xlabel('Class')
ax.set_title('V1 Probability Density Distribution')
ax.legend(loc = 'best')

plt.show()
```
